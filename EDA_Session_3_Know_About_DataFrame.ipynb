$take$

- take is used to extract the particular index of data

- take has 1 parameter called axis

- axis=1 represents columns

- axis=0 represents rows

- by default axis=0 is there

- In below example we given 100,200,300 and we did not provided axis value

- It understood we are taking about rows

```python
visa_df.take([100,200,300])
```

```python
visa_df.take([100,200,300], axis=1) # we dont have columns with 100, 200 and 300 so its not working
```

```python
visa_df.take([2,5,7], axis=1) # it is giving columns
```

```python
visa_df.take([100:105]) # not working, slicing not work with take
```

```python
# Q)
# If I want rows 100 200 300 and columns : 2,5,7
# with single syntax (with single take) it is not possible
# So that is the drawback of take
# Slicing is not poosible in take
```

```python
visa_df.take([100,200,300]).take([2,5,7],axis=1)
```

$iloc$

**visa_df.iloc[<rows>,<columns>]**

**visa_df.iloc[<start:end>,<start:end>]**

- we can overcome drawback of take with iloc

- We can get both rows and columns retrieval at a time is possible

- **Note**

    - No bracket, then it is series

    - with bracket then dataframe

```python
visa_df.iloc[[100,200,300],[2,5,7]]
```

```python
visa_df.iloc[[100,200,300]] # same like take
```
